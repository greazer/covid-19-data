```python
import github as gh

this_milestone = 'August 2021'

# Create instance using an access token
g = gh.Github("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")

# Connect to interesting repos
repo_jupyter = g.get_organization('microsoft').get_repo('vscode-jupyter')
repo_vscode = g.get_organization('microsoft').get_repo('vscode')
repo_jupyter_internal = g.get_organization('microsoft').get_repo('vscode-jupyter-internal')
repo_vscode_internalbacklog = g.get_organization('microsoft').get_repo('vscode-internalbacklog')

repos_all = [repo_jupyter, repo_vscode, repo_jupyter_internal, repo_vscode_internalbacklog]
```

```python

# Deal with rate limiting by filtering by current milestone
repo_ms_nums = {}
for r in repos_all:
    milestones = r.get_milestones()
    for m in milestones:
        if m.title == this_milestone:
            repo_ms_nums[r.name] = r.get_milestone(m.number)
            break

if len(repo_ms_nums) != len(repos_all):
    raise Exception('One repo is missing milestone: {}'.format(this_milestone))
```

```python
# Pull in all issues into a list
issues_list = []

for r in repos_all:
   for issue in r.get_issues(state='open', milestone=repo_ms_nums[r.name]):
      issues_list.append([r.name, issue.number, issue.title, 'https://github.com/microsoft/{0}/issues/{1}'.format(r.name, issue.number), 'Nobody' if type(issue.assignee) is type(None) else issue.assignee.name, issue.labels])
```

```python
# Create a pandas dataframe for all issues

import pandas as pd
repo_df = pd.DataFrame(issues_list, columns=['repo', 'number', 'title', 'link','assignee', 'labels'])

def make_clickable(val):
    return '<a href="{}">link</a>'.format(val)

repo_df.style.format({'link':make_clickable})
```
